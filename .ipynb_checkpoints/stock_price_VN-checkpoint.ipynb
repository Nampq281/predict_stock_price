```python
import pandas as pd
```

```python
import glob
```

```python
data_path = 'dataset'
```

```python
data_list = glob.glob(f'{data_path}/*.csv')
```

```python
data_list
```

```python
result = []
for i in data_list:
    result.append(pd.read_csv(i))

df = pd.concat(result)
```

```python
df['Ticker'].value_counts()
```

```python
df.info()
```

```python
t = df['Date/Time'].iloc[0]
```

```python
from datetime import datetime as dt
def format_time(row):
    row = dt.strptime(row, '%m/%d/%Y %H:%M')
    return row

df['time'] = df['Date/Time'].apply(lambda row: format_time(row))
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df_plot = df[df['Ticker']=='FPT']
```

```python
# Plot Closing Prices
plt.figure(figsize=(14, 7))
sns.lineplot(data=df_plot, x=df_plot.time, y="Close", label="Close", color='green')
plt.title("Stock Closing Prices Over Time")
plt.xlabel("Date/Time")
plt.ylabel("Price ($)")
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Plot Volume (Bar Plot)
plt.figure(figsize=(14, 7))
sns.barplot(data=df_plot, x=df_plot.time, y="Volume", color='gray')
plt.title("Trading Volume Over Time")
plt.xlabel("Date/Time")
plt.ylabel("Volume")
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
